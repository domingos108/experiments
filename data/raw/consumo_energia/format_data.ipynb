```python
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
import numpy as np
```

```python
serie_name = 'consumosulformated.txt'
df = pd.read_csv(serie_name)
df['y'] = np.log(df['y'])
df.to_csv(serie_name.replace('.txt', 'log.txt'), index=False)
```

```python
df['y'].plot()
```

```python
plot_acf(df['y'], lags=20, use_vlines=True, title='Autocorrelation Function (ACF)')
plt.show()
plot_pacf(df['y'], lags=20, use_vlines=True, title='Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
df['y'].diff().dropna()
```

```python
plot_acf(df['y'].diff().dropna(), lags=20, use_vlines=True, title='Autocorrelation Function (ACF)')
plt.show()
plot_pacf(df['y'].diff().dropna(), lags=20, use_vlines=True, title='Partial Autocorrelation Function (PACF)')
plt.show()
```
