```python
from pathlib import Path

import config
from utils.compare_fs_vs_baseline import build_comparison
from utils.export_metrics_to_csv import save_csv

%load_ext autoreload
%autoreload 2
```

```python
# === CELULA DE CONFIGURACAO -- mesmo padrao de compare_fs_results_mlp.ipynb/
# compare_fs_results_svr.ipynb ===
# Versao para a familia ARIMA-SVR (hibrido) -- baseline_dir continua SEMPRE
# data/result/chamados, modelo baseline '1as'. linear_model_name_to_exclude
# ='1arima' porque o .pkl do ARIMA copiado vive dentro de cada pasta de FS
# (Additive precisa dele sob o mesmo experiment_id) -- mesmo padrao da
# familia hibrida ARIMA-MLP (compare_fs_results.ipynb), diferente das
# familias single-model (MLP/SVR), que nao copiam nada.
baseline_dir = Path(config.MODEL_DATA_PATH) / 'chamados'
baseline_model_name = '1as'
linear_model_name_to_exclude = '1arima'

# (rotulo, experiment_id) de cada variante de FS a comparar -- rotulo vira o
# prefixo das colunas no DataFrame de saida (<rotulo>_RMSE, <rotulo>_PctGain,
# <rotulo>_NFeatures).
fs_variants = [
    ('ftest', 'chamados_v4_fs_arimasvr_ftest'),
    ('mutualinfo', 'chamados_v4_fs_arimasvr_mutualinfo'),
    ('rfembedded', 'chamados_v4_fs_arimasvr_rfembedded'),
    ('lasso', 'chamados_v4_fs_arimasvr_lasso'),
    ('rfecv', 'chamados_v4_fs_arimasvr_rfecv'),
]

output_path = Path(config.ROOT_PATH) / 'results' / 'chamados_v4_fs_arimasvr_comparison.csv'
```

```python
# Compara o baseline x cada variante de FS lado a lado, por serie -- mesma
# funcao usada pela CLI (src/utils/compare_fs_vs_baseline.py) e pelos
# notebooks irmaos das outras familias, so o baseline_model_name/
# linear_model_name_to_exclude mudam (celula 1).
fs_dirs = {rotulo: Path(config.MODEL_DATA_PATH) / experiment_id for rotulo, experiment_id in fs_variants}

df_comparison = build_comparison(
    baseline_dir,
    fs_dirs,
    baseline_model_name=baseline_model_name,
    linear_model_name_to_exclude=linear_model_name_to_exclude,
)
save_csv(df_comparison, output_path, label='CSV de comparacao ARIMA-SVR (baseline x variantes FS)')
df_comparison
```
