```python
from pathlib import Path

import config
from utils.compare_fs_vs_baseline import build_comparison
from utils.export_metrics_to_csv import save_csv

%load_ext autoreload
%autoreload 2
```

```python
# === CELULA DE CONFIGURACAO -- Tarefa 6-gate/7-gate (mesmo padrao de
# notebook/compare_fs_results_mlp.ipynb, Tarefa 6-gate) ===
# Versao para a familia SVR single (SKlearnModel) -- baseline_dir continua
# SEMPRE data/result/chamados (os 5 baselines intocaveis da Secao 3 do
# CLAUDE.md), mas o MODELO baseline dentro dele e '1svr'.
baseline_dir = Path(config.MODEL_DATA_PATH) / 'chamados'
baseline_model_name = '1svr'
# SVR single (SKlearnModel) nao copia nenhum modelo linear pre-treinado para
# dentro do result_dir de FS -- None desliga esse filtro por completo.
linear_model_name_to_exclude = None

# (rotulo, experiment_id) de cada variante de FS a comparar -- rotulo vira o
# prefixo das colunas no DataFrame de saida (<rotulo>_RMSE, <rotulo>_PctGain,
# <rotulo>_NFeatures).
fs_variants = [
    ('ftest', 'chamados_v4_fs_svr_ftest'),
    ('mutualinfo', 'chamados_v4_fs_svr_mutualinfo'),
    ('rfembedded', 'chamados_v4_fs_svr_rfembedded'),
    ('lasso', 'chamados_v4_fs_svr_lasso'),
    ('rfecv', 'chamados_v4_fs_svr_rfecv'),
]

output_path = Path(config.ROOT_PATH) / 'results' / 'chamados_v4_fs_svr_comparison.csv'
```

```python
# Compara o baseline x cada variante de FS lado a lado, por serie -- mesma
# funcao usada pela CLI (src/utils/compare_fs_vs_baseline.py) e pelos
# notebooks irmaos das outras familias (compare_fs_results.ipynb,
# compare_fs_results_mlp.ipynb), so o baseline_model_name/
# linear_model_name_to_exclude mudam (celula 1).
fs_dirs = {rotulo: Path(config.MODEL_DATA_PATH) / experiment_id for rotulo, experiment_id in fs_variants}

df_comparison = build_comparison(
    baseline_dir,
    fs_dirs,
    baseline_model_name=baseline_model_name,
    linear_model_name_to_exclude=linear_model_name_to_exclude,
)
save_csv(df_comparison, output_path, label='CSV de comparacao SVR single (baseline x variantes FS)')
df_comparison
```
