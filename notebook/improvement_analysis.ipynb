```python
experiment_id = 'residualens'
```

```python
df_mean_metrics, df_all_metrics, df_prevs = metrics.open_fold_result(experiment_id)
```

```python
df_results = df_mean_metrics.reset_index().sort_values(by=['ts', 'model'])[
    ['ts', 'model', 'MSE' ,'RMSE',	'MAPE', 'MAE']
]
```

```python
base = ['1arima', '1aeml'] 
competidor = '1ammean100'
metric_base = 'RMSE'
list_compare_metrics = []
for key, df_ts in df_results.groupby(['ts']):
    base_metrics_filtered = df_ts[df_ts['model'] == base][metric_base]
    base_metric = base_metrics_filtered.iloc[0]
    if base_metrics_filtered.shape[0]>1:
        raise Exception(f'{key} com multiplos baselines ')
    compare_metrics = df_ts.set_index('model')[metric_base] < base_metric
    list_compare_metrics.append(compare_metrics.to_dict())
```
