```python
import warnings
from statsmodels.tools.sm_exceptions import InterpolationWarning

warnings.simplefilter('ignore', InterpolationWarning)

from sklearn.neural_network import MLPRegressor

from model import perturbative, grid_search_exp

import config
%load_ext autoreload
%autoreload 2
```

```python
experiment_id = 'pertexps'
normalize = None

force = True
model_exec = 10

model_parameters = {
    'hidden_layer_sizes': [10, 15, 20, 50], 
    'max_iter': [1000],
}
```

```python
for bag_size in [1]:
    for qtd_pert in [10, 20, 30, 40, 50]:
        for lr in [0.1, 0.5, 0.01]:
            model_name = f'mlpb{bag_size}{qtd_pert}{lr}'
            model = MLPRegressor(activation='relu', solver='sgd') 
            
            experiment_params = {
                'models_before': None,
                "model_pertub": MLPRegressor(activation='relu', solver='sgd'),
                "lag_size": None,
                "learning_rate_init" : lr,
                "qtd_perturbations": qtd_pert,
                'earlystop': None,
                'bagging_pct': bag_size
                
            }
            grid_search_exp.grid_seach_multiple_bases(
                 perturbative.Perturbative, model, normalize, model_parameters,experiment_params,
                 model_exec, model_name, experiment_id, force = force
            )
```
