```python
import warnings
from statsmodels.tools.sm_exceptions import InterpolationWarning

warnings.simplefilter('ignore', InterpolationWarning)

from sklearn.neural_network import MLPRegressor

from model import perturbative, grid_search_exp

import config
%load_ext autoreload
%autoreload 2
```

```python
experiment_id = 'pertexps'
normalize = None

force = True
model_exec = 10

model_parameters = {
    'hidden_layer_sizes': [2, 5, 10, 15, 20, 50], 
    'max_iter': [1000],
}
```

```python
for lr in [0.6,0.8]:
    for qtd_pert in [100]:
        model_name = f'mlpwk{lr}{qtd_pert}5'
        model = MLPRegressor(activation='logistic', solver='lbfgs') 
        
        experiment_params = {
            'models_before': None,
            "model_pertub": MLPRegressor(activation='logistic', solver='lbfgs'),
            "lag_size": None,
            "learning_rate_init" :lr ,
            "qtd_perturbations": qtd_pert,
            'earlystop': 5
            
        }
        grid_search_exp.grid_seach_multiple_bases(
             perturbative.Perturbative, model, normalize, model_parameters,experiment_params,
             model_exec, model_name, experiment_id, force = force
        )
```
