```python
import warnings
from statsmodels.tools.sm_exceptions import InterpolationWarning

warnings.simplefilter('ignore', InterpolationWarning)

from sklearn.svm import SVR

from model import perturbative, grid_search_exp

import config
%load_ext autoreload
%autoreload 2
```

```python
experiment_id = 'pertexps'
normalize = None

force = True
model_exec = 1

model_parameters =  {
    'C':[10, 100, 1000], 'gamma': ['auto'],
    'kernel':["rbf"],
    'epsilon': [0.1, 0.01, 0.001], 'tol':[0.001],
}
```

```python
for lr in [0.6,0.8, 1]:
    for qtd_pert in [50]:
        model_name = f'svrwk{lr}{qtd_pert}5'
        model = SVR(max_iter=100000)
        
        experiment_params = {
            'models_before': None,
            "model_pertub":SVR(max_iter=100000),
            "lag_size": None,
            "learning_rate_init" :lr ,
            "qtd_perturbations": qtd_pert,
            'earlystop': 1,
            'diff_kpss':True
            
        }
        grid_search_exp.grid_seach_multiple_bases(
             perturbative.Perturbative, model, normalize, model_parameters,experiment_params,
             model_exec, model_name, experiment_id, force = force
        )
```
