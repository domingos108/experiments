```python
from sklearn.svm import SVR

from model import generics, perturbative


import config
%load_ext autoreload
%autoreload 2
```

```python
base_name_list = ['lynx.txt','sunspot.txt']
lag_size_list = [5, 18]
normalize=True

model_exec = 1
experiment_id = 'test'
model_name = 'pertp2'
models_before = 'pertp1'
```

```python
from sklearn.model_selection import ParameterGrid

for base_name,lag_size in zip(base_name_list, lag_size_list):

    fold, title = generics.format_names(experiment_id, base_name, models_before)
    pn=generics.open_saved_result(title)[0]['params']
    
    fixed = pn.get('model_params', [])
    if len(fixed) ==0:
        fixed = [pn]

    variavel = list(ParameterGrid({'C':[10, 100, 1000], 'gamma': [0.1, 0.01, 0.001],
                      'kernel':["rbf"],
                      'epsilon': [0.1, 0.01, 0.001], 'tol':[0.001]
                     }
                 ))
    model = perturbative.Perturbative(model_base = SVR(max_iter=100000))

    parameters = {
    'model_params': [ fixed+[v] for v in variavel],
        'model_base': [SVR(max_iter=100000)]
    
    }

    print(base_name)
    generics.grid_seach(model,  base_name, 
               normalize, 
               lag_size, 
               parameters, model_exec,
               model_name, experiment_id)
```
