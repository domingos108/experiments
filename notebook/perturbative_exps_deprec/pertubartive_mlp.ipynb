```python
import warnings
from statsmodels.tools.sm_exceptions import InterpolationWarning

warnings.filterwarnings("ignore", category=UserWarning)
warnings.simplefilter('ignore', InterpolationWarning)

from sklearn.neural_network import MLPRegressor

from model import generics, perturbative
from sklearn.model_selection import ParameterGrid

import config
%load_ext autoreload
%autoreload 2
```

```python
base_name = 'mlpgs'
model_name_list = [
    f'{base_name}p1', 
    f'{base_name}p2', 
    f'{base_name}p3', 
]
model_before_list = [
    base_name, 
    f'{base_name}p1', 
    f'{base_name}p2', 
]

model_base = MLPRegressor(activation='logistic', solver='lbfgs') 

params_modelo_base = list(
    ParameterGrid( 
        {'hidden_layer_sizes': [2, 5, 10, 15, 20, 50], 
        'max_iter': [1000],
        'time_window': [12]
        }
))
```

```python
base_name_list = config.BASE_NAME_LIST
normalize=False
force = False
model_exec = 10
experiment_id = 'pertexps'

for model_name, models_before in zip(
    model_name_list, 
    model_before_list
):
    
    perturbative.exec_model(
        base_name_list, 
        experiment_id, 
        models_before,
        model_base, 
        params_modelo_base,
        normalize,
        model_exec,
        model_name,
        force
    )
    
```
