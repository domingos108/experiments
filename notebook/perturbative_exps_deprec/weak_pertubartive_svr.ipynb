```python
import warnings
from statsmodels.tools.sm_exceptions import InterpolationWarning

warnings.filterwarnings("ignore", category=UserWarning)
warnings.simplefilter('ignore', InterpolationWarning)

from sklearn.svm import SVR

from model import generics, perturbative
from sklearn.model_selection import ParameterGrid

import config
%load_ext autoreload
%autoreload 2
```

```python
grid_params = list(ParameterGrid( 
        {
        'C':[10, 100, 1000],#, 100, 1000],
        'gamma': ['auto'],#[0.1, 0.01, 0.001],
        'kernel':["rbf"],
        'epsilon': [0.001, 0.1 , 0.01], 
        'tol':[0.001],
        'time_window': [2, 6, 8, 12]
    }
))
```

```python
for lr_type in ['constant', 'invscaling' ]:
    for pert in [4, 6, 8, 10]:
        for lr  in [0.5, 1, 0.1]:
            base_name_list = config.BASE_NAME_LIST
            normalize=False
            force = False
            diff_kpss = False
            model_exec = 1
            parameters = {
                'model_params': [[c]for c in grid_params],
                'qtd_pertubacao': [pert],
                'model_base': [SVR(max_iter=100000)],
                'learning_rate': [lr_type],
                'learning_rate_init': [lr],
                'fixed_error_lags':[True]
            }
            
            model = perturbative.Perturbative(model_base = SVR(max_iter=100000)) 
            model_name = f'svrv1{lr_type}{lr}{pert}'
            experiment_id = 'test'
            
            generics.grid_seach_multiple_bases(model,  base_name_list, normalize,
                                               parameters, model_exec, model_name, experiment_id, force, diff_kpss)
```
