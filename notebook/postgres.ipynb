```python
import psycopg2
from psycopg2 import Error
import os

def list_all_postgres_tables():
    """
    Conecta-se ao banco de dados PostgreSQL e lista todas as tabelas
    no esquema 'public'. As credenciais são lidas de variáveis de ambiente.
    """
    connection = None
    try:
        # Recupera as credenciais do banco de dados de variáveis de ambiente
        # Isso é uma boa prática para não "hardcodar" credenciais no código.
        db_name = os.environ.get('DJANGO_DB_NAME', 'inagency')
        db_user = os.environ.get('DJANGO_DB_USER', 'admininagency')
        db_password = os.environ.get('DJANGO_DB_PASSWORD', 'KjjKdX@231pq')
        db_host = os.environ.get('DJANGO_DB_HOST', '18.218.6.51')
        db_port = os.environ.get('DJANGO_DB_PORT', '5432')


        print(f"Tentando conectar ao PostgreSQL em {db_host}:{db_port}...")

        # Estabelece a conexão com o PostgreSQL
        connection = psycopg2.connect(
            user=db_user,
            password=db_password,
            host=db_host,
            port=db_port,
            database=db_name
        )
        cursor = connection.cursor()

        print("Conexão ao PostgreSQL bem-sucedida!")
        print("\nListando tabelas no esquema 'public':")

        # Query SQL para obter os nomes das tabelas no esquema 'public'
        # information_schema.tables é uma visão padrão SQL que fornece informações sobre tabelas e vistas
        # table_schema = 'public' filtra para o esquema padrão
        # table_type = 'BASE TABLE' filtra apenas para tabelas (exclui views)
        cursor.execute("""
            SELECT table_name
            FROM information_schema.tables
            WHERE table_schema = 'public'
            AND table_type = 'BASE TABLE';
        """)

        tables = cursor.fetchall()

        if tables:
            for table in tables:
                print(f"- {table[0]}")
        else:
            print("Nenhuma tabela encontrada no esquema 'public'.")

    except (Exception, Error) as error:
        print(f"Erro ao conectar ou consultar o PostgreSQL: {error}")
    #finally:
    # Fecha a conexão com o banco de dados
    #if connection:
    # cursor.close()
    # connection.close()
    #print("Conexão ao PostgreSQL fechada.")

    return cursor
```
