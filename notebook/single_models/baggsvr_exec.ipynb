```python
import os
os.environ['PYTHONWARNINGS'] = 'ignore'

import warnings
from statsmodels.tools.sm_exceptions import InterpolationWarning
from sklearn.exceptions import ConvergenceWarning

warnings.simplefilter('ignore', InterpolationWarning)
warnings.filterwarnings("ignore", category=ConvergenceWarning)

from sklearn.svm import SVR
from sklearn.ensemble import BaggingRegressor

from model import single_ml_model_exp, grid_search_exp

import config
%load_ext autoreload
%autoreload 2


for n_estimators in [50]:
    for max_samples in [0.8]:
        for max_features in [0.8]:
            model = BaggingRegressor(n_jobs=-1)
            
            experiment_id = 'irradiacao2024'
            model_name = f'svrbagg{n_estimators}{max_samples}{max_features}'
            normalize = True
            force = True
            model_exec = 10
            
            experiment_params = {
                'diff_kpss': False,
                'horizon': 1
            }
            
            model_parameters = {
                        'estimator': [
                            SVR(max_iter=100000, C=10, epsilon = 0.1, gamma='auto', kernel='rbf', tol=0.001),
                            SVR(max_iter=100000, C=100, epsilon = 0.1, gamma='auto', kernel='rbf', tol=0.001),
                            SVR(max_iter=100000, C=1000, epsilon = 0.1, gamma='auto', kernel='rbf', tol=0.001),
                            SVR(max_iter=100000, C=10, epsilon = 0.01, gamma='auto', kernel='rbf', tol=0.001),
                            SVR(max_iter=100000, C=100, epsilon = 0.01, gamma='auto', kernel='rbf', tol=0.001),
                            SVR(max_iter=100000, C=1000, epsilon = 0.01, gamma='auto', kernel='rbf', tol=0.001),
                            SVR(max_iter=100000, C=10, epsilon = 0.001, gamma='auto', kernel='rbf', tol=0.001),
                            SVR(max_iter=100000, C=100, epsilon = 0.001, gamma='auto', kernel='rbf', tol=0.001),
                            SVR(max_iter=100000, C=1000, epsilon = 0.001, gamma='auto', kernel='rbf', tol=0.001),
                                     ], 
                        'n_estimators': [n_estimators],
                        'max_features': [max_features],
                        'max_samples': [max_samples]
                        
                    }
    
            grid_search_exp.grid_seach_multiple_bases(
                 single_ml_model_exp.SKlearnModel, model, normalize, model_parameters,experiment_params,
                 model_exec, model_name, experiment_id, force = force
            )
```
