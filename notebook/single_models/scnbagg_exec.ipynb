```python
import warnings
from statsmodels.tools.sm_exceptions import InterpolationWarning
warnings.simplefilter('ignore', InterpolationWarning)

from model.scn import SCN_III, sigmoid
from sklearn.ensemble import BaggingRegressor

from model import single_ml_model_exp, grid_search_exp

import config
%load_ext autoreload
%autoreload 2


for n_estimators in [10, 20 , 50, 100]:
    for max_samples in [0.6, 0.8]:
        model = BaggingRegressor(n_jobs=-1)
        
        experiment_id = 'consumoenergia'
        model_name = f'sncbagg{n_estimators}{max_samples}'
        normalize = True
        force = False
        model_exec = 10
        
        experiment_params = {
            'diff_kpss': True,
            "lag_size": None
        }
        
        model_parameters = {
                    'estimator': [SCN_III(hidden_act_fuction=sigmoid, t_max=10, l_max=5 ) 
                                 ], 
                    'n_estimators': [n_estimators],
                    'max_features': [1],
                    'max_samples': [max_samples]
                    
                }

        grid_search_exp.grid_seach_multiple_bases(
             single_ml_model_exp.SKlearnModel, model, normalize, model_parameters,experiment_params,
             model_exec, model_name, experiment_id, force = force
        )
```
