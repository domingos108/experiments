```python
import warnings
from statsmodels.tools.sm_exceptions import InterpolationWarning

warnings.filterwarnings("ignore", category=UserWarning)
warnings.simplefilter('ignore', InterpolationWarning)
from model.elm import Model as ELM
from sklearn.neural_network import MLPRegressor

from model import generics, perturbative
from sklearn.model_selection import ParameterGrid

import config
%load_ext autoreload
%autoreload 2
```

```python
base_name = 'elm'

model_base = ELM() 
model_pert = MLPRegressor(activation='logistic', solver='lbfgs')

params_modelo_base = list(
    ParameterGrid({'hidden_layer_sizes': [2, 5, 10, 15, 20, 50], 
        'max_iter': [1000],
        'time_window': [2, 6, 8, 12]
        }
))
```

```python
base_name_list = config.BASE_NAME_LIST
normalize=False
force = True
model_exec = 10
experiment_id = 'test'
for lr in ['constant']:
    for qtd_pertubacao in  [4, 6, 8, 10]:
        for learning_rate_init in [1]:
        
            model_name_list = [f'{base_name}mlp{lr}{qtd_pertubacao}{learning_rate_init}'.replace('.', '')]
            model_before_list = [base_name]
            
            for model_name, models_before in zip(
                model_name_list, 
                    model_before_list
            ):
                
                perturbative.exec_model(
                    base_name_list, 
                    experiment_id, 
                    models_before,
                    model_base, 
                    params_modelo_base,
                    normalize,
                    model_exec,
                    model_name,
                    force = force,
                    qtd_perturbative = qtd_pertubacao,
                    learning_rate_init = learning_rate_init,
                    fixed_error_lags = True,
                    learning_rate = lr,
                    model_pert = model_pert
                )
            
```
