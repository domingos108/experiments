```python
import warnings
from statsmodels.tools.sm_exceptions import InterpolationWarning

warnings.filterwarnings("ignore", category=UserWarning)
warnings.simplefilter('ignore', InterpolationWarning)
from sklearn.svm import SVR

from model import generics, perturbative
from sklearn.model_selection import ParameterGrid

import config
%load_ext autoreload
%autoreload 2
```

```python
base_name = 'svrdiff'

model_base = SVR(max_iter=100000)
params_modelo_base = list(
    ParameterGrid({
        'C':[10, 100, 1000],
        'gamma': ['auto'],
        'kernel':["rbf"],
        'epsilon': [0.1, 0.01, 0.001], 
        'tol':[0.001],
        'time_window': [2, 6, 8, 12]
    }
))
```

```python
base_name_list = config.BASE_NAME_LIST
normalize=False
force = False
model_exec = 1
experiment_id = 'test'
lr = 'constant'

for qtd_pertubacao in  [4, 6, 8 , 10]:
    for learning_rate_init in [1, 0.5, 0.1]:
    
        model_name_list = [f'kpss{base_name}{lr}{qtd_pertubacao}{learning_rate_init}'.replace('.', '')]
        model_before_list = [base_name]
        
        for model_name, models_before in zip(
            model_name_list, 
                model_before_list
        ):
            
            perturbative.exec_model(
                base_name_list, 
                experiment_id, 
                models_before,
                model_base, 
                params_modelo_base,
                normalize,
                model_exec,
                model_name,
                force = force,
                qtd_perturbative = qtd_pertubacao,
                learning_rate_init = learning_rate_init,
                fixed_error_lags = True,
                learning_rate = lr
            )
        
```
