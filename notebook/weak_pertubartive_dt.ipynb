```python
import warnings
from statsmodels.tools.sm_exceptions import InterpolationWarning

warnings.filterwarnings("ignore", category=RuntimeWarning)

warnings.filterwarnings("ignore", category=UserWarning)
warnings.simplefilter('ignore', InterpolationWarning)
from sklearn.tree import DecisionTreeRegressor

from model import generics, perturbative
from sklearn.model_selection import ParameterGrid

import config
%load_ext autoreload
%autoreload 2

```

```python
grid_params = list(ParameterGrid( 
        {
        'max_depth': [5, 10, 15],     # Maximum depth of the tree
        'min_samples_split': [2, 5],   # Minimum samples required to split an internal node
        'min_samples_leaf': [1, 2],      # Minimum samples required to be at a leaf node
        'max_features': ['sqrt'],
        'time_window': [2, 6, 8, 12] # Number of features to consider when looking for the best split
    }
))
```

```python
for pert, lr in zip(
    [5, 10],
    [0.1, 0.1]
):
    base_name_list = config.BASE_NAME_LIST
    normalize=False
    force = False
    
    model_exec = 1
    parameters = {
        'model_params': [[c]for c in grid_params],
        'qtd_pertubacao': [pert],
        'model_base': [DecisionTreeRegressor()],
        'learning_rate': ['constant'],
        'learning_rate_init': [lr],
        'fixed_error_lags':[True]
    }
    
    model = perturbative.Perturbative(model_base = DecisionTreeRegressor()) 
    model_name = f'dtweakhigh{lr}{pert}'
    experiment_id = 'test'
    
    generics.grid_seach_multiple_bases(model,  base_name_list, normalize,
                                       parameters, model_exec, model_name, experiment_id, force)
```
